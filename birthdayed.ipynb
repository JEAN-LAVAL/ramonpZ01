```python
#birthday.py
import datetime
import calendar 

def Birthday_day(yar,mothn,dae):
   # daynum=date(yer,mothn,dae).isoweekday()
   # dayweek=%a
    dayweek=calendar.weekday(yar,mothn,dae)
    return dayweek

# Prompt a user to enter the month and day of their birthday, and print what day of
# the week their birthday falls on this year.
ya=int(input("Please enter year in which you want the day: "))
mnoth=str(input("Please enter birth month in letters: "))
if mnoth.lower()=="january":
    mnth=1
if mnoth.lower()=="february":
    mnth=2
if mnoth.lower()=="march":
    mnth=3
if mnoth.lower()=="april":
    mnth=4
if mnoth.lower()=="may":
    mnth=5
if mnoth.lower()=="june":
    mnth=6
if mnoth.lower()=="july":
    mnth=7
if mnoth.lower()=="august":
    mnth=8
if mnoth.lower()=="september":
    mnth=9
if mnoth.lower()=="october":
    mnth=10
if mnoth.lower()=="november":
    mnth=11
if mnoth.lower()=="december":
    mnth=12

dya=int(input("Please enter the day you were born in number: "))
reqday=Birthday_day(ya,mnth,dya)
if reqday==0:
    bday="Monday"
if reqday==1:
    bday="Tuesday"
if reqday==2:
    bday="Wednesday"
if reqday==3:
    bday="Thursday"
if reqday==4:
    bday="Friday"
if reqday==5:
    bday="Saturday"
if reqday==6:
    bday="Sunday"
print(f"Your birthday falls on a {bday} in {ya}")

```
